## Consider only top 100 for matched names

```python
def limit_rows(group, max_rows=100):
    return group.head(max_rows)

c_df = c_df.groupby(['name']).apply(limit_rows).reset_index(drop=True)
```

## Find Edit distance

```python
import Levenshtein

# Define a function to calculate Levenshtein distance
def edit_distance(row):
    return Levenshtein.distance(row['password_x'], row['password_y'])

# Apply the function to 'password_x' and 'password_y' to get 'edit_distance'
c_df['edit_distance'] = c_df.apply(edit_distance, axis=1)
```

```python
c_df[['name' ,'password_x','password_y','edit_distance']][c_df['name'] == '11xxaidanxx11@yahoo.com'].head(50)
```

```python
# Group by 'name' and calculate the min, avg edit distance
min_avg_edit_distance = c_df.groupby('name')['edit_distance'].agg(['min', 'mean']).reset_index()

# Fetch the password_x that has the minimum edit_distance
best_password_df = c_df.loc[c_df.groupby('name')['edit_distance'].idxmin(),['name', 'password_x', 'password_y']]

# Merge min, avg edit distance dataframe with best_password dataframe
final_df = pd.merge(min_avg_edit_distance, best_password_df, on="name")
```
